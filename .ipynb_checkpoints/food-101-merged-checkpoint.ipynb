## Food-101 dataset

The Food-101 dataset has 101 food categories, with 1000 images (512x512) provided for each category for a total of 101,000 images. The Food-101 dataset provides a list of examples for training as well as another list for testing. Since the dataset specifies an equal amount of examples for each category, we do not need to worry about class imbalances.The dataset is split into 20% for the validation set and 80% for the training set.

Training set: 80800 images

Validation set: 20200 images

The validation set is used to check the performance of the model after each training epoch.

```python
# Input root directory
input_root_dir = "../input/food-101/food-101/food-101"
input_root_path = Path(input_root_dir)
print(os.listdir(input_root_dir))
```

```python
# Image directory
image_dir_path = input_root_path/'images'
```

```python

src = (ImageList.from_folder(image_dir_path)
       .split_by_rand_pct(0.2)       
       .label_from_folder()) 
```

## Image Pre-processing

We do **data augmentation** to increase the diversity of the data available for training models without collection of new data. The images are resized to 256x256 for faster training of the model. Data augmentation during training uses the set of transforms provided by the Fast AI library and some extra transformations as shown in the next cell. These are provided by the get_transforms() method. The pixel values of the examples are also normalized based on the ImageNet values. These examples could have been normalized by determining the mean & standard deviation of this specific dataset, but it appears that the ImageNet values were sufficient for a good result.

```python
tfms = get_transforms(max_rotate=10,
                      max_zoom=1.1,
                      max_lighting=None,
                      max_warp=0.2,
                      xtra_tfms=[
                          brightness(change=(0.5-0.2, 0.5+0.2), p=0.75),
                          contrast(scale=(1-0.4, 1+0.2), p=0.75),
                          squish(scale=(1-0.3, 1+0.5), p=0.75),
                          skew(direction=(0, 7), magnitude=random.randint(0,6)/10, p=0.75),]
                      )

# batch size
bs = 64
```

```python
# create databunch

data = (src.transform(tfms, size=256)  #resize images to 256
        .databunch(bs=bs)              #batch size=64
        .normalize(imagenet_stats))
```

Let's see the dataset.

```python
data.show_batch(rows=3, figsize=(9, 7))
```

```python

print(data.classes)
len(data.classes)
```

## Model Definition

Here we are using ResNet-50 as our model architechture. We will be using transfer learning approaches to train our model.

### Why ResNet?

ResNet is a short name for Residual Network. In general, in a deep convolutional neural network, several layers are stacked and are trained to the task at hand. The network learns several level features at the end of its layers. 

In residual learning, instead of trying to learn some features, we try to learn some residual. Residual can be simply understood as subtraction of feature learned from input of that layer. ResNet does this using shortcut connections (directly connecting input of n<sup>th</sup> layer to some (n+x)<sup>th</sup> layer.)

![Residual Block Image](https://i.imgur.com/L82YBFq.png)

It has proved that training this form of networks is easier than training simple deep convolutional neural networks and also the problem of degrading accuracy is resolved.


If we see in the previous SoTA results, we can see that there was no need for complex network as vanilla ResNet50 pre-trained on Imagenet provided good results. So, we also choose the same pre-trained ResNet50 as our starting point.

**Weight Decay:**
We also use weight decay, which is yet another regularization technique which prevents the weights from becoming too large by adding an additional term to the loss function.

```python
# Defining the CNN Model

food_learn = cnn_learner(
    data,
    models.resnet50,  # Resnet 50 architechture
    metrics=accuracy,
    wd=1e-1,         # weight decay
)

# Change the output directory to the current working directory
food_learn.model_dir = os.getcwd()
food_learn.path = Path(os.getcwd())

# Display model architecture
food_learn.model
```

```python
food_learn.summary()
```

## Stage - 1 

Training the model first with with freezed layers. In Fast AI library when we load Resnet-50 or any other arcitechture, they have freezed layers by default. You can see that the Trainable parameters in the cell above are set to False indicating that the layers are freezed.

### Choosing Learning Rate

Now we come to the problem of what learning rate to choose. We need to find a learning rate so that our model reaches the minimum loss within a reasonable amount of time. if we take the learning rate too small, then the train time will be much higher and if we take it too high, then we will not be able to reach the minimal loss.

![Learning rates](https://www.jeremyjordan.me/content/images/2018/02/Screen-Shot-2018-02-24-at-11.47.09-AM.png)

We could simply choose a lower learning rate and see the accuracy and loss statistics and tune it ourselves but that may take a long time to accomplish a good learning rate. However, we don't need to do that since the FastAI library that we are using provides a method called `lr_find()` which does a small run where the learning rate is increased after each processed batch and the corresponding loss is logged. The result of this is a lr vs. loss plot that can be used as guidance for choosing a optimal initial learning rate.

```python
# Finding a good learning rate

food_learn.lr_find()
food_learn.recorder.plot()
```

### Learning rate scheduling

Instead of using a fixed learning rate, we will use a learning rate scheduler, which will change the learning rate after every batch of training. There are many strategies for varying the learning rate during training, and the one we'll use is called the **"One Cycle Learning Rate Policy"**, which involves starting with a low learning rate, gradually increasing it batch-by-batch to a high learning rate for about 30% of epochs, then gradually decreasing it to a very low value for the remaining epochs. 

```python
# learning rate
lr = 1e-2

# Train with 1 cycle policy
food_learn.fit_one_cycle(10, slice(lr))

# save the model 
food_learn.save('256_stage1', return_path=True)

```

```python
# Plot loss history
food_learn.recorder.plot_losses()
```

```python
food_learn.recorder.plot_metrics()
```

```python
# Check the final validation accuracy
food_learn.validate(food_learn.data.valid_dl)
```

We see that after training out model with freezed layers for 10 epochs resulted in a 77.3% accuracy. Now, its time to unfreeze the model and train it.

```python
# Plot loss history
food_learn.recorder.plot_losses()
```

The case of Validation Loss lower than training Loss as shown in the graph above, we see that validation loss is now lower than training loss.


Reasons why this might be happening : 

* The training loss is calculated over the entire training dataset. Likewise, the validation loss is calculated over the entire validation dataset. The training set is typically at least 4 times as large as the validation (80-20). Given that the error is calculated over all samples, you could expect up to approximately 4X the loss measure of the validation set. [[Stackoverflow Answer](https://stats.stackexchange.com/questions/187335/validation-error-less-than-training-error)]

* Regularization applied during training, but not during validation. 

<blockquote class="twitter-tweet"><p lang="en" dir="ltr">The most common reason is regularization (e.g., dropout), since it applies during training, but not during validation &amp; testing. * Due to Data Augmentation or Validation Set easier than Train Set
